```python
import pandas as pd
from pathlib import Path

# Load data
PROJECT_ROOT = Path.cwd()
DATA_PATH = PROJECT_ROOT / "data" / "processed" / "data.parquet"
if not DATA_PATH.exists():
    DATA_PATH = PROJECT_ROOT.parent / "data" / "processed" / "data.parquet"

df = pd.read_parquet(DATA_PATH)

# Explicitly choose correct date column
date_col = "date"

# Ensure datetime (already is, but safe)
df[date_col] = pd.to_datetime(df[date_col], errors="coerce")

print(f"Using date column: {date_col}")
print(f"Null ratio: {df[date_col].isna().mean():.2%}")

# ✅ Final answer to Question 5
print("Time coverage:")
print(df[date_col].min(), "→", df[date_col].max())
```
